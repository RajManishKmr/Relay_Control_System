```python
Time = []
for i in range(len(A)):
    wcr = (2*3.1415)/(Ti[i])
    a = (4*H[i])/((3.1415)*A[i])
    b = a**2 - 1
    T = ((b**0.5)/wcr).real
    Time.append(T) 
```

```python
Delay = []
for i in range(len(A)):
    
    wcr = (2*3.1415)/(Ti[i]*3)
    a = (4*H[i])/((3.1415)*A[i])
    b = a**2 - 1
    arr = [b**0.5]
    d = np.arctan(arr)
    Delay.append((3.1415 - d[0])/wcr)
    
```
